```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.io import wavfile
import scipy.io as sci
```

```python
# ..........Plot Wave 1..........
sample_rate1, data_wav1 = wavfile.read("Waves/wave1.wav")

time1 = np.linspace(0, len(data_wav1)/sample_rate1, len(data_wav1))

plt.plot(time1[:500], data_wav1[:500])
plt.show()
```

```python
def integrate(a, b , x, t):
    sumX = 0
    n = b - a
    
    for i in range(a+1,b):

        sumX += x[i]

    sumX += x[a]/2 + x[b]/2

    return sumX * (t[b] - t[a]) / n
```

```python
m = 11

# ..........Computing Fourier Series Coefficients Wave 1..........
t = 101
plt.plot(time1[:t], data_wav1[:t])
plt.show()

x1 = time1[:t]
y1 = data_wav1[:t]
L1 = x1[t - 1]
```

```python
# Fourier Coefficients
ak1 = np.zeros(m, dtype=complex)

ak1[0] = 2./L1 * integrate(0, len(x1) - 1, y1, x1)
an1 = lambda n: 2.0/L1 * integrate(0, len(x1) - 1, y1 * np.cos(2.*np.pi*n*x1/L1), x1)
bn1 = lambda n: 2.0/L1 * integrate(0, len(x1) - 1, y1 * np.sin(2.*np.pi*n*x1/L1), x1)

for i in range(1,m):
    ak1[i] = complex(an1(i), -1 * bn1(i))
```

```python
t1 = np.arange(0, len(ak1) * 1/L1, 1/L1)
plt.stem(t1, abs(ak1))
plt.show()
```

```python
# Series sum
terms = 50
s1 = ak1[0]/2. + sum([an1(k) * np.cos(2.*np.pi*k*x1/L1) + bn1(k) * np.sin(2.*np.pi*k*x1/L1) for k in range(1,terms+1)])

# Plotting the Fourier Series Signal via Original
plt.plot(x1[1:100], s1.real[1:100], label="Fourier series")
plt.plot(x1, y1, label="Original wave")
plt.xlabel("$x$")
plt.ylabel("$y=f(x)$")
plt.legend(loc='best',prop={'size':10})
plt.title("Triangular wave signal analysis by Fouries series")
plt.show()
```

```python
# امتیازی 2
# ...............Bonus Wave...............
bonusRate, BonusWave = wavfile.read("Waves/bonus.wav")

timeB = np.linspace(0, len(BonusWave)/bonusRate, len(BonusWave))

plt.plot(timeB[:500], BonusWave[:500])
plt.show()
```

```python
# ..........Computing Fourier Series Coefficients Wave 1..........
Lb = 0.22934

tb = 0
for i in range(10000):
    g = BonusWave[i]
    if abs(g - Lb) <= 0.001:
        tb = i
        break

plt.plot(timeB[:tb], BonusWave[:tb])
plt.show()

xb = timeB[:tb]
yb = BonusWave[:tb]
```

```python
# Fourier Coefficients
m = 150
akB = np.zeros(m, dtype=complex)

akB[0] = 2./Lb * integrate(0, len(xb) - 1, yb, xb)
anB = lambda n: 2.0/Lb * integrate(0, len(xb) - 1, yb * np.cos(2.*np.pi*n*xb/Lb), xb)
bnB = lambda n: 2.0/Lb * integrate(0, len(xb) - 1, yb * np.sin(2.*np.pi*n*xb/Lb), xb)

for i in range(1,m):
    akB[i] = complex(anB(i), -1 * bnB(i))
```

```python
# Coefficients Plot
tB = np.arange(0, len(akB) * 1/Lb, 1/Lb)
plt.stem(tB, abs(akB))
plt.show()
```

```python
# Series sum
terms = 150
sb = akB[0]/2. + sum([anB(k) * np.cos(2.*np.pi*k*xb/Lb) + bnB(k) * np.sin(2.*np.pi*k*xb/Lb) for k in range(1,terms+1)])

# Plotting the Fourier Series Signal via Original
plt.plot(xb, sb.real, label="Fourier series")
plt.plot(xb, yb, label="Original wave")
plt.xlabel("$x$")
plt.ylabel("$y=f(x)$")
plt.legend(loc='best',prop={'size':10})
plt.title("Triangular wave signal analysis by Fouries series")
plt.show()
```
